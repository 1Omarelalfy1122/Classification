```python
import pandas as pd
```

```python
data = pd.read_csv("cars.csv")
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data.columns
```

```python
y= data['buying']
```

```python
X = data [['maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.25 , random_state= 44 , shuffle = True)
```

```python
from sklearn.ensemble import RandomForestClassifier
RandomForestModell = RandomForestClassifier(criterion='gini',max_depth=2,n_estimators = 100,random_state = 33)
```

```python
categorical_cols = X_train.select_dtypes(include=['object']).columns
print(categorical_cols)
X_train_encoded = pd.get_dummies(X_train, columns=categorical_cols)
X_test_encoded = pd.get_dummies(X_test, columns=categorical_cols)

X_test_encoded = X_test_encoded.reindex(columns=X_train_encoded.columns, fill_value=0)


RandomForestModell.fit(X_train_encoded,y_train)
```

```python
y_pred = RandomForestModell.predict(X_test_encoded)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report
```

```python
Report = classification_report(y_test,y_pred)
```

```python
print('Report : \n',Report)
```
